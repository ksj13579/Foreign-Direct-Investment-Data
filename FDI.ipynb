```python
years=['2000-01', '2001-02', '2002-03', '2003-04', '2004-05',
       '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11',
       '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']
```

```python
#Creating the Average Exchange Rate List
rate=[45.68,47.69,48.39,45.95,44.93,44.27,45.24,48.26,45.99,47.44,45.56,47.92,
      54.48,60.50,61.14,65.46,67.87]
```

### Convert the Currency US$ to INR

```python
#Convert FDI value from US$ to INR
def multiply_columns(fdi, col_list,num):
    for col in col_list:
        fdi[col] = fdi[col]*rate[col_list.index(col)]/10
    return fdi
```

```python
fdi_inusd=fdi.copy()
fdi2 = multiply_columns(fdi, years, rate)
```

```python
#FDI inflows (Amount in Crores)
fdi2.style.set_caption('Amounts in crores').format(precision=2)
```

```python
#Unpivoting melt DataFrame
melt= pd.melt(fdi_inusd, id_vars = ['Sector'], value_vars=years, var_name='years',
             value_name='fdi(usd)', ignore_index=True)
melt
```

```python
melt01= pd.melt(fdi2, id_vars = ['Sector'], value_vars=years, var_name='years',
             value_name='fdi(Crores)', ignore_index=True)
melt01=round(melt01,2)
melt01
```

```python
#Merging the FDI (US$ Million) column of melt DataFrame into melt01 DataFrame
merged=melt01.merge(melt,how='left')
merged
```

```python
#Sorting the sectors and year columns
sorted = merged.sort_values(['Sector','years'],ignore_index=True)
sorted
```

```python
print('\n state for sectors \n','-'*70, sep='')
print(pd.DataFrame(sorted.groupby('Sector').describe().loc[:,:]).transpose())
```

```python
#Replace long values of sector column to short form
sorted = sorted[['Sector','fdi(Crores)','fdi(usd)']].replace(['CONSTRUCTION DEVELOPMENT:township, housing, built-up infastructure and construction-development project',
                                                             'service sector(fin, banking, insurance, Non fin/Business, outsourcing, R&D, Courier, Tech. testig and analysis,other)',
                                                             'tea and coffee (processing and marehousing coffee & rubber)'],
                                                             ['construction development','services sector','tea and coffee'])
```

## Sector-wise FDI 2000-2017:

```python
#Groupby sector column to find total FDI inflow per sector from 2000-01 to 2016-17 
Sectorwise_fdi = sorted.groupby('Sector').sum()
Sectorwise_fdi.sort_values(by='fdi(usd)',ascending=False)
```

## Visualization:

```python
Sectorwise_fdi.plot(kind='bar',y='fdi(Crores)', figsize=(25,7), legend= True, title='sector_wise fdi inflows', ylabel='fdi(crore)')
plt.show()
```

```python
Sectorwise_fdi.plot(kind='bar',y='fdi(usd)', figsize=(25,7), legend= True, title='sector_wise fdi inflows', ylabel='fdi(usd)')
plt.show()
```

## Top 10 Sectors

```python
top_10_sectors = Sectorwise_fdi.nlargest(10,['fdi(Crores)'])
```

```python
Total_fdi = round(melt01['fdi(Crores)'].sum(),2)
sum = top_10_sectors['fdi(Crores)'].sum()
top_10_sectors['in %age']=round(top_10_sectors['fdi(Crores)']/sum*100,2)
top_10_sectors['%age to total inflows']=round(top_10_sectors['fdi(Crores)']/sum*100,2)
top_10_sectors
```

## Visualization

```python
plt.figure(figsize=(15,5))
plt.barh(top_10_sectors.index,top_10_sectors['fdi(Crores)'])
plt.title('sector attracting highest FDI inflows')
plt.xlabel('fdi(Crores)')
plt.ylabel('top sectors')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.pie(top_10_sectors['fdi(Crores)'], labels=top_10_sectors.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Sharae Among Top 10 Sectors Attracting Highest FDI Inflows')
plt.show()
```

## Year_wise FDI Inflow

```python
#create DataFrame
melt02=melt01[['years','fdi(Crores)']]
melt02=round(melt02.groupby('years').sum(),2)
```

```python
#creating new column of % growth over previous year 
melt02['%age growth over previous year']=round(melt02.pct_change()*100,2)
```

```python
print('*'*8,'Details of Variation of FDI Inflow Year_wise','*'*8 )
melt02.fillna('-')
```

```python
#plotting to show year by year FDI inflow
melt02.plot.line(y='fdi(Crores)',figsize=(15,15))
plt.ylabel('fdi(Crores)')
plt.title('FDI by year inflow')
plt.show()
```

### The above graph shows the total amount of FDI inflows in india during the last 17 years i.e. 2000 - 2017 


*The FDI inflows from 2000-2001 i.e Rs. 10865.97Cr. In 2001-2002 it was Rs. 19208.02 Cr. It shows that the good result in FDI inflows in india. Little bit ups and downs in FDi inflow upto 2005-2006, but after that great hike in the year of 2007-2008 and FDI was Rs. 118600.87Cr. In 2008-2009 ther was a huge investment in FDI in Rs.144390.03Cr, but then there was a downfall in inflow of FDI in two successive years 2009-2010 and 2010-2011, with figure Rs. 122558.27Cr and Rs.97421.29Cr respectively. We can analysis from the graph that in the year 2011-2012 the inflow of the FDI was second highest of the last fifteen years
i.e. Rs.168298.80Cr. Year 2012-2013 and 2013-2014 the FDI inflow fluctuated Rs. 122163.67Cr and Rs. 147010.90Cr. respectively. In the last financial year i.e.2016-2017 the amount of the FDI inflow were Rs.295086.94 Cr. which is the highest FDI inflows in last 17 years.
